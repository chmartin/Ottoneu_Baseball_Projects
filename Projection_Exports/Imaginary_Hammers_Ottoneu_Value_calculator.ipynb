# Notebook to join Custom $Values for export suitable for the Ottoneu Value Calculator

```python
League_path = '/Users/cmartin/Fantasy_Baseball/Ottoneu_Baseball_Projects/Imaginary_Hammers/'
```

```python
Proj_files = {
    'Hitter':'Latest_Hitter_Proj.csv',
    'Pitcher':'Latest_Pitcher_Proj.csv'
}
```

```python
Player_ID_Map_Path = '/Users/cmartin/Fantasy_Baseball/Ottoneu_Baseball_Projects/Player_ID_Map/Latest_Player_ID_Map.csv'
```

```python
import pandas as pd
```

```python
Export_Cols = {
    'OTTONEUID':'otto',
    'FANGRAPHSNAME':'Name',
    'POS':'Pos',
    'TEAM':'Team',
    'TOTAL_SGP_Val':'$'
}
```

```python
Player_ID_Map_df = pd.read_csv(Player_ID_Map_Path)
```

```python
Hitter_Proj_df = pd.read_csv(League_path+Proj_files['Hitter'])
Pitcher_Proj_df = pd.read_csv(League_path+Proj_files['Pitcher'])
```

```python
Hitter_Proj_df.head()
```

```python
Player_ID_Map_df[Player_ID_Map_df['FANGRAPHSNAME'] == 'Mike Trout'].iloc[0]
```

```python
Hitter_Proj_df.rename(columns={
    'PlayerId':'IDFANGRAPHS'
},inplace=True)
Pitcher_Proj_df.rename(columns={
    'PlayerId':'IDFANGRAPHS'
},inplace=True)
```

```python
Hitter_merge = Hitter_Proj_df[['IDFANGRAPHS','TOTAL_SGP_Val']].merge(Player_ID_Map_df[['IDFANGRAPHS','OTTONEUID','POS','FANGRAPHSNAME','PLAYERNAME','TEAM']],how='inner')
```

```python
Pitcher_merge = Pitcher_Proj_df[['IDFANGRAPHS','TOTAL_SGP_Val']].merge(Player_ID_Map_df[['IDFANGRAPHS','OTTONEUID','POS','FANGRAPHSNAME','PLAYERNAME','TEAM']],how='inner')
```

```python
Hitter_merge[(Hitter_merge['FANGRAPHSNAME'].isna()) & (Hitter_merge['TOTAL_SGP_Val'] > -1.)]['TOTAL_SGP_Val'].count()
```

```python
Pitcher_merge[(Pitcher_merge['FANGRAPHSNAME'].isna()) & (Pitcher_merge['TOTAL_SGP_Val'] > -1.)]['TOTAL_SGP_Val'].count()
```

```python
Hitter_merge[(Hitter_merge['TOTAL_SGP_Val'] >= 1.)]['TOTAL_SGP_Val'].sum()
```

```python
Pitcher_merge[(Pitcher_merge['TOTAL_SGP_Val'] >= 1.)]['TOTAL_SGP_Val'].sum()
```

```python
Hitter_Pitcher = pd.concat(
    [
        Hitter_merge,
        Pitcher_merge
    ]
)
```

```python
Hitter_Pitcher.head()
```

```python
Final_df = Hitter_Pitcher.rename(columns=Export_Cols)[[val for key,val in Export_Cols.items()]]
```

```python
Final_df.head()
```

```python
Final_df.to_csv(League_path+'Ottoneu_Value_Calc_Custom$Values.csv',index=False)
```
