# Notebook to join Custom $Values for export suitable for the Ottoneu Value Calculator

```python
League_path = '/Users/cmartin/Fantasy_Baseball/Ottoneu_Baseball_Projects/Imaginary_Hammers/'
```

```python
Proj_files = {
    'Hitter':'Latest_Hitter_Proj.csv',
    'Pitcher':'Latest_Pitcher_Proj.csv'
}
```

```python
Player_ID_Map_Path = '/Users/cmartin/Fantasy_Baseball/Ottoneu_Baseball_Projects/Player_ID_Map/Latest_Player_ID_Map.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
Export_Cols = {
    'Ottoneu ID':'otto',
    'Name':'Name',
    'Ottoneu Positions':'Pos',
    'Team':'Team',
    'TOTAL_SGP_Val':'$'
}
```

```python
Player_ID_Map_df = pd.read_csv(Player_ID_Map_Path)
```

```python
Hitter_Proj_df = pd.read_csv(League_path+Proj_files['Hitter'])
Pitcher_Proj_df = pd.read_csv(League_path+Proj_files['Pitcher'])
```

```python
Hitter_Proj_df.rename(columns={
    'PlayerId':'FG ID'
},inplace=True)
Pitcher_Proj_df.rename(columns={
    'PlayerId':'FG ID'
},inplace=True)
```

```python
Hitter_Proj_df[Hitter_Proj_df['Name'] == 'Sebastian Walcott'][['FG ID','TOTAL_SGP_Val','Team']].reset_index(drop=True).iloc[0]
```

```python
Hitter_Proj_df[Hitter_Proj_df['Name'] == 'Shohei Ohtani'][['FG ID','TOTAL_SGP_Val','Team']].reset_index(drop=True).iloc[0]
```

```python
Player_ID_Map_df['FG ID'] = Player_ID_Map_df['FG ID'].astype(str)
```

```python
Player_ID_Map_df['FG ID'] = Player_ID_Map_df['FG ID'].fillna(Player_ID_Map_df['FG Minor ID'])
```

```python
Player_ID_Map_df[Player_ID_Map_df['Name'] == 'Sebastian Walcott'].reset_index(drop=True).iloc[0]
```

```python
Player_ID_Map_df[Player_ID_Map_df['Name'] == 'Shohei Ohtani'].reset_index(drop=True).iloc[0]
```

```python
Player_ID_Map_df['FG ID'] = Player_ID_Map_df['FG ID'].str.replace('.0','')
```

```python
Hitter_merge = Hitter_Proj_df[['FG ID','TOTAL_SGP_Val','Team','POS']].merge(Player_ID_Map_df[['FG ID','Ottoneu ID','Ottoneu Positions','Name']],how='inner')
```

```python
Pitcher_merge = Pitcher_Proj_df[['FG ID','TOTAL_SGP_Val','Team','POS']].merge(Player_ID_Map_df[['FG ID','Ottoneu ID','Ottoneu Positions','Name']],how='inner')
```

```python
#Handle Othtani...
Ohtani_Value = Hitter_Pitcher[Hitter_Pitcher['FG ID'] == '19755']['TOTAL_SGP_Val'].sum()
Ohtani_df = pd.DataFrame([
    {
        'FG ID':'19755',
        'TOTAL_SGP_Val':Ohtani_Value,
        'Team':'LAD',
        'POS':"['DH','SP']",
        'Ottoneu ID':33600,
        'Ottoneu Positions':'Util/SP',
        'Name':'Shohei Ohtani'
    }
])
```

```python
Hitter_Pitcher.drop_duplicates(['FG ID'],keep=False,inplace=True)
```

```python
Hitter_Pitcher = pd.concat([
    Hitter_Pitcher,
    Ohtani_df
])
```

```python
Hitter_Pitcher[Hitter_Pitcher['FG ID'].isin([key for key, val in Hitter_Pitcher['FG ID'].value_counts().to_dict().items() if val > 1])].sort_values('Name')
```

```python
Final_df = Hitter_Pitcher.rename(columns=Export_Cols)[[val for key,val in Export_Cols.items()]]
```

```python
Final_df = Final_df.sort_values('$',ascending=False).reset_index(drop=True)
```

```python
Final_df.to_csv(League_path+'Ottoneu_Value_Calc_Custom$Values.csv',index=False)
```
