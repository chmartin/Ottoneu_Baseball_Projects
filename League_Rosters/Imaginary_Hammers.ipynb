```python
def symlink_force(target, link_name):
    try:
        os.symlink(target, link_name)
    except OSError as e:
        if e.errno == errno.EEXIST:
            os.remove(link_name)
            os.symlink(target, link_name)
        else:
            raise e
```

```python
Hitter_latest_link_name = League_path+'/Current_League_Rosters.csv'
symlink_force(f'{Write_Path}/{League_Name}_{export_date}.csv',Hitter_latest_link_name)
```
