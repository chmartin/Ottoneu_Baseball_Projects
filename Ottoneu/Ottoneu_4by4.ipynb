# Download GameType Averages

```python
League_path = '/Users/cmartin/Fantasy_Baseball/Ottoneu_Baseball_Projects/Ottoneu'
```

```python
export_date = 'Jan26_2026'
```

```python
GameType = 1
GameName = 'Ottoneu_4x4'
```

```python
url = f'https://ottoneu.fangraphs.com/averageValues?export=csv&gameType={GameType}'
```

```python
import pandas as pd
```

```python
Ottoneu_df = pd.read_csv(url)
```

```python
Ottoneu_df.head()
```

```python
import os
import errno
Write_Path = League_path+'/'+GameName
if not os.path.exists(Write_Path):
    os.makedirs(Write_Path)
```

```python
Ottoneu_df.to_csv(f'{Write_Path}/{export_date}.csv')
```

```python
def symlink_force(target, link_name):
    try:
        os.symlink(target, link_name)
    except OSError as e:
        if e.errno == errno.EEXIST:
            os.remove(link_name)
            os.symlink(target, link_name)
        else:
            raise e
```

```python
latest_link_name = League_path+f'/Latest_{GameName}.csv'
symlink_force(f'{Write_Path}/{export_date}.csv',latest_link_name)
```
